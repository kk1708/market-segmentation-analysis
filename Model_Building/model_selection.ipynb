# Model Selection

We are dealing with a classification problem. To solve a classification problem, there are various classification algorithms that are used. Some of the popular ones are:
- `Logistic Regression`
- `K-Nearest Neighbours`
- `Random Forest Classifier`
- `Support Vector Machines`

Depending on our dataset, some algorithms work better than others. Our job is finding the best set of algorithms. Finding these algorithms require a good understanding of the dataset. This understanding comes from **visualising the data in an effective manner**

```python
# importing the required libraries

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# loading the dataset and observing a few values

df = pd.read_csv('../dataset/Train_After.csv')
df.sample(10)
```

## Plotting a Pair Plot

A pair plot allows us to observe the relationship between all points of the dataset. Based on this visualisation, a suitable learning algorithm is used

```python
sns.pairplot(df.drop(columns='ID'), hue='Segmentation')
```

Based on the Pair Plot, we make the following observations:
- In some cases, there are no overlaps
- Some of the cases have overlaps, but not too much
- The datapoints are arranged in straight lines

## Algorithm Selection

**The given dataset requires a multiclass learning algorithm because each element belongs to just one class.** Using this knowledge in conjunction with the observations made above, we decide on a set of classification algorithms, which will give us the best results. A few popular multiclass algorithms are:
- `K-Nearest Neighbours`
- `Random Forest`
- `Decision Trees`
- `Naive Bayes`
- `Gradient Boosting`

Out of these algorithms, we will try `KNN`, `Random Forest` and `Gradient Boosting`
- `KNN` is a good algorithm for data points which overlap each other
- `Random Forest` is a good algorithm when points are easily seperable. In our dataset, some cases have no overlaps and are linear in nature. This allows the use of decision tree based algorithms. The only drawback is the time taken to train the model
- `Gradient Boosting` is a boosting technique that uses decision trees. The idea here is to optimise the loss function. The disadvantage is the computational cost required. [Learn more here about GB](https://www.youtube.com/watch?v=Nol1hVtLOSg)
